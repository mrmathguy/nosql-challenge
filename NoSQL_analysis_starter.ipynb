### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

#So there are no LocalAuthorityName called London, however, London has a bunch of local authorities

#So I'm going to use beautiful soup to scrape a list of local authorities in London I found of Wikipedia
from splinter import Browser
from bs4 import BeautifulSoup
browser = Browser('chrome')


url = "https://en.wikipedia.org/wiki/List_of_London_boroughs"
browser.visit(url)
html=browser.html
soup = BeautifulSoup(html,'html.parser')

tr =soup.find_all('tbody')[0].find_all("tr")
london_authorities = []

for row in tr:
    london_authorities.append(row.td.find_all('a')[0].text)



query = {"$and":[{"LocalAuthorityName":{"$in":london_authorities}}, {'RatingValue': {"$gte":4}}]}


london=db.establishment.find(query)
print(db.establishment.count_documents(query))
# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint
pprint(db.establishment.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
# Display the number of rows in the DataFrame
london_pd = pd.DataFrame(list(london))
# Display the first 10 rows of the DataFrame
london_pd.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
query={"BusinessName":"Penang Flavours"}
penang = db.establishment.find_one(query)
lat = float(penang['geocode']['latitude'])
lon = float(penang['geocode']['longitude'])
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = lat
longitude = lon

query = {'$and':[{"geocode.latitude":{"$lte":latitude+degree_search}},
                 {"geocode.latitude": {'$gte': latitude-degree_search}},
                {"geocode.longitude":{"$lte":longitude+degree_search}},
                 {"geocode.longitude": {'$gte': longitude-degree_search}}, 
                {'RatingValue':{'$eq':5}}]}
sort =  [("score.Hygiene", 1)]

business_around_Pentang  = db.establishment.find(query).sort(sort)
for business in business_around_Pentang:
    pprint(business)
# Print the results
```

```python
# Convert result to Pandas DataFrame
#print(business_around_Pentang)
business_around_Pentang = db.establishment.find(query).sort(sort)
businesses_near_Pentang_pd = pd.DataFrame(list(business_around_Pentang))
businesses_near_Pentang_pd 
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
sort= [("LocalAuthorityName", -1)]
query = {"scores.Hygiene":{'$eq':0}}






# Print the number of documents in the result
print(db.establishment.count_documents(query))

# Print the first 10 results
pprint(list(db.establishment.find(query).sort(sort).limit(10)))
```

```python
# Convert the result to a Pandas DataFrame
bad_hygiene_pd = pd.DataFrame(list(db.establishment.find(query)))
# Display the number of rows in the DataFrame
print(len(bad_hygiene_pd))
# Display the first 10 rows of the DataFrame
bad_hygiene_pd.head(10)
```
